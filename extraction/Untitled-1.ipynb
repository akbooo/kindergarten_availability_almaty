```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def parse():
    driver = webdriver.Chrome()

    all_data = []

    for start in range(0, 730, 5):  # 40 страниц × 5
        url = f"https://open-almaty.kz/obrazovatelnye-uchrezhdeniya/detskie-sady-chastnye?start={start}"
        driver.get(url)

        time.sleep(2)

        items = driver.find_elements(By.CLASS_NAME, "spoiler-wrap")

        for item in items:
            try:
                name = item.find_element(By.CLASS_NAME, "spoiler-head").text

                # кликаем, чтобы раскрыть
                driver.execute_script(
                    "arguments[0].click();",
                    item.find_element(By.CLASS_NAME, "spoiler-head")
                )
                time.sleep(0.2)

                # адрес
                try:
                    address = item.find_element(
                        By.XPATH, ".//span[text()='Адрес:']/following-sibling::span"
                    ).text
                except:
                    address = None

                # телефон
                try:
                    phone = item.find_element(
                        By.XPATH, ".//span[contains(text(),'Контактная информация:')]/following-sibling::span"
                    ).text
                except:
                    phone = None
                try:
                    type = item.find_element(
                        By.XPATH, ".//span[text()='Тип учреждения:']/following-sibling::span"
                    ).text
                except:
                    type = None

                try:
                    email = item.find_element(
                        By.XPATH, ".//span[text()='Email:']/following-sibling::span"
                    ).text
                except:
                    email = None

                all_data.append({
                    "name": name,
                    "address": address,
                    "phone": phone,
                    "type": type,
                    "email": email
                })

            except Exception as e:
                print("Ошибка:", e)

        print(f"Готово: start={start}")

    driver.quit()
    return all_data


data = parse()
print(len(data))
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.to_csv("C:\\Users\\Subbota\\kinder\\kindergartens_almaty_private.csv", index=False, encoding="utf-8-sig")

print("Сохранено в CSV")
```
